```python
import matplotlib.pyplot as plt

# Data
Temperatures = [5, 10, 40, 60, 80, 100, 150, 200, 289]
resonance_diff = [0.13, 0.13, 0.15, 0.15, 0.15, 0.17, 0.18, 0.22, 0.38]


# Plotting
plt.figure(figsize=(10, 6))

plt.plot(Temperatures, resonance_diff, marker='o', label='Resonance Difference', linestyle='-')

# Labels and title
plt.xlabel("Temperature (K)", fontsize=12)
plt.ylabel("Resonance Difference", fontsize=12)
#plt.title("Resonance Differences vs Temperature", fontsize=14)
plt.legend()
#plt.grid(True, linestyle='--', alpha=0.6)

# Annotations
ax = plt.gca()
ax.annotate(r'$YMn_6Sn_6$', ha='left', xy=(0.9, 0.97), weight='bold', xycoords='axes fraction', fontsize=10, color='black')
ax.annotate(r'$\nu=120\ GHz$', ha='left', xy=(0.9, 0.9), xycoords='axes fraction', fontsize=10, color='black')
#ax.annotate(r'$B_\perp$', ha='left', xy=(0.9, 0.86), xycoords='axes fraction', fontsize=22)
#ax.annotate(r'$B_\parallel$', ha='left', xy=(0.9, 0.87), xycoords='axes fraction', fontsize=22)

# Display plot
plt.tight_layout()
plt.savefig("C:\\Users\\lovia\\Desktop\\YMnSn_120_55_GHz_Resonance difference.png", dpi=200, bbox_inches='tight')
plt.show()
```
