```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data based on the description
# Temperature range
temperature = np.linspace(100, 300, 200)

# In-plane data
in_plane_g = 2.15 + 0.2 * np.sin((temperature - 100) / 200 * np.pi)
in_plane_linewidth = 5 + 6 * np.exp(-((temperature - 250) ** 2) / (2 * 30 ** 2))

# Out-of-plane data
out_of_plane_g = 2.0 - 0.4 * np.sin((temperature - 100) / 200 * np.pi)
out_of_plane_linewidth = 20 + 50 * np.exp(-((temperature - 250) ** 2) / (2 * 30 ** 2))

# Plotting
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# g-factor plot
axs[0].plot(temperature, in_plane_g, label='In-Plane g-factor')
axs[0].plot(temperature, out_of_plane_g, label='Out-of-Plane g-factor')
#axs[0].set_title('g-factor vs Temperature')
axs[0].set_xlabel('Temperature (K)')
axs[0].set_ylabel('g-factor')
axs[0].legend()

# Linewidth plot
axs[1].plot(temperature, in_plane_linewidth, label='In-Plane Linewidth')
axs[1].plot(temperature, out_of_plane_linewidth, label='Out-of-Plane Linewidth')
#axs[1].set_title('Linewidth vs Temperature')
axs[1].set_xlabel('Temperature (K)')
axs[1].set_ylabel('Linewidth (10^-2 T)')
axs[1].legend()

plt.tight_layout()
plt.show()
```
