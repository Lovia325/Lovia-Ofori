```python
# Calculate and histogram the longitudinal and transverse momenta 
#  of the first and second leptons in each event.
#  use 50 bins between 0 and 500 GeV

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
Events = np.loadtxt("Events.txt")

# Define column names
columns = ['event_number', 'run_number', 
           'particle1_Px', 'particle1_Py', 'particle1_Pz', 'particle1_E', 'particle1_charge', 'particle1_ID',
           'particle2_Px', 'particle2_Py', 'particle2_Pz', 'particle2_E', 'particle2_charge', 'particle2_ID']

# Convert the NumPy array to a Pandas DataFrame
df = pd.DataFrame(Events, columns=columns)

# Calculate longitudinal momentum (Pz) for each lepton
df['particle1_Pz_long'] = df['particle1_Pz']
df['particle2_Pz_long'] = df['particle2_Pz']

# Calculate transverse momentum (Pt) for each lepton
df['particle1_Pt'] = np.sqrt(df['particle1_Px']**2 + df['particle1_Py']**2)
df['particle2_Pt'] = np.sqrt(df['particle2_Px']**2 + df['particle2_Py']**2)

# Create histograms for longitudinal and transverse momenta
plt.figure(figsize=(12, 6))

# Longitudinal momentum histograms
plt.subplot(1, 2, 1)
plt.hist(df['particle1_Pz_long'], bins=50, range=(0, 500), alpha=0.5, label='Highest pt')
plt.hist(df['particle2_Pz_long'], bins=50, range=(0, 500), alpha=0.5, label='Second pt')
plt.xlabel('pL/GeV')
plt.ylabel('Number of Events')
plt.legend()
plt.title('Longitudinal Momentum of Leptons')

# Transverse momentum histograms
plt.subplot(1, 2, 2)
plt.hist(df['particle1_Pt'], bins=50, range=(0, 500), alpha=0.5, label='Highest pt')
plt.hist(df['particle2_Pt'], bins=50, range=(0, 500), alpha=0.5, label='Second pt')
plt.xlabel('pT/GeV')
plt.ylabel('Number of Events')
plt.legend()
plt.title('Transverse Momentum of Leptons')

plt.tight_layout()
plt.show()
```

```python
# Plot the transverse momentum of lepton 1 vs the transverse momentum
#  of lepton 2.  Use scales 0 to 500 GeV on both axes.

# Plot transverse momentum of lepton 1 vs transverse momentum of lepton 2
plt.figure(figsize=(7, 5))
plt.scatter(df['particle1_Pt'], df['particle2_Pt'], alpha=0.5)
plt.xlim(0, 500)
plt.ylim(0, 500)
plt.xlabel('pT of Lepton 1 (GeV)')
plt.ylabel('pT of Lepton 2 (GeV)')
plt.title('Transverse Momentum of Lepton 1 vs Lepton 2')
plt.grid(False)
plt.show()
```

```python
# Repeat the above two steps for your saved sub-samples from last time.
#  Do you notice anything interesting?

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
Events = np.loadtxt("Events.txt")

# Define column names
columns = ['event_number', 'run_number', 
           'particle1_Px', 'particle1_Py', 'particle1_Pz', 'particle1_E', 'particle1_charge', 'particle1_ID',
           'particle2_Px', 'particle2_Py', 'particle2_Pz', 'particle2_E', 'particle2_charge', 'particle2_ID']

# Convert the NumPy array to a Pandas DataFrame
df = pd.DataFrame(Events, columns=columns)

# Calculate transverse momentum (pt) for each lepton
df['particle1_Pt'] = np.sqrt(df['particle1_Px']**2 + df['particle1_Py']**2)
df['particle2_Pt'] = np.sqrt(df['particle2_Px']**2 + df['particle2_Py']**2)

# Define longitudinal momentum as Pz for each lepton
df['particle1_Pz_long'] = df['particle1_Pz']
df['particle2_Pz_long'] = df['particle2_Pz']

# Separate same-sign and different-sign events
same_sign = df[df['particle1_charge'] == df['particle2_charge']]
different_sign = df[df['particle1_charge'] != df['particle2_charge']]

# Function to create histograms for longitudinal and transverse momenta
def plot_momenta(df, title_prefix):
    plt.figure(figsize=(12, 6))

    # Longitudinal momentum histograms
    plt.subplot(1, 2, 1)
    plt.hist(df['particle1_Pz_long'], bins=50, range=(0, 500), alpha=0.5, label='Particle 1')
    plt.hist(df['particle2_Pz_long'], bins=50, range=(0, 500), alpha=0.5, label='Particle 2')
    plt.xlabel('Longitudinal Momentum (GeV)')
    plt.ylabel('Number of Events')
    plt.legend()
    plt.title(f'{title_prefix}: Longitudinal Momentum of Leptons')

    # Transverse momentum histograms
    plt.subplot(1, 2, 2)
    plt.hist(df['particle1_Pt'], bins=50, range=(0, 500), alpha=0.5, label='Particle 1')
    plt.hist(df['particle2_Pt'], bins=50, range=(0, 500), alpha=0.5, label='Particle 2')
    plt.xlabel('Transverse Momentum (GeV)')
    plt.ylabel('Number of Events')
    plt.legend()
    plt.title(f'{title_prefix}: Transverse Momentum of Leptons')

    plt.tight_layout()
    plt.show()

# Plot momenta for same-sign events
plot_momenta(same_sign, "Same-Sign")

# Plot momenta for different-sign events
plot_momenta(different_sign, "Different-Sign")

# Function to create scatter plots for transverse momenta
def plot_scatter(df, title_prefix):
    plt.figure(figsize=(8, 6))
    plt.scatter(df['particle1_Pt'], df['particle2_Pt'], alpha=0.5)
    plt.xlim(0, 500)
    plt.ylim(0, 500)
    plt.xlabel('Transverse Momentum of Lepton 1 (GeV)')
    plt.ylabel('Transverse Momentum of Lepton 2 (GeV)')
    plt.title(f'{title_prefix}: Transverse Momentum of Lepton 1 vs Lepton 2')
    plt.grid(False)
    plt.show()

# Plot scatter for same-sign events
plot_scatter(same_sign, "Same-Sign")

# Plot scatter for different-sign events
plot_scatter(different_sign, "Different-Sign")
```
