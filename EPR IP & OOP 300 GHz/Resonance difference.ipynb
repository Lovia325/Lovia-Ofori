```python
import matplotlib.pyplot as plt

# Data
in_plane_temperatures = [100, 150, 200, 290]
out_of_plane_temperatures = [100, 125, 160, 200, 240, 285]
in_plane_resonance_diff = [1.3, 1.21, 1.13, 0.95]
out_of_plane_resonance_diff = [-1.77, -1.7, -1.59, -1.47, -1.3, -1.1]

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(in_plane_temperatures, in_plane_resonance_diff, marker='o', label='In-Plane Resonance Difference', linestyle='-')
plt.plot(out_of_plane_temperatures, out_of_plane_resonance_diff, marker='s', label='Out-of-Plane Resonance Difference', linestyle='--')

# Labels and title
plt.xlabel("Temperature (K)", fontsize=12)
plt.ylabel("Resonance Difference", fontsize=12)
#plt.title("Resonance Differences vs Temperature", fontsize=14)
plt.legend()
#plt.grid(True, linestyle='--', alpha=0.6)

# Annotations
ax = plt.gca()
ax.annotate(r'$YMn_6Sn_6$', ha='left', xy=(0.9, 0.8), weight='bold', xycoords='axes fraction', fontsize=10, color='black')
ax.annotate(r'$\nu=300\ GHz$', ha='left', xy=(0.9, 0.7), xycoords='axes fraction', fontsize=10, color='black')
#ax.annotate(r'$B_\perp$', ha='left', xy=(0.9, 0.86), xycoords='axes fraction', fontsize=22)
#ax.annotate(r'$B_\parallel$', ha='left', xy=(0.9, 0.87), xycoords='axes fraction', fontsize=22)

# Display plot
plt.tight_layout()
plt.savefig("C:\\Users\\lovia\\Desktop\\YMnSn_300GHz_Resonance difference.png", dpi=200, bbox_inches='tight')
plt.show()
```
